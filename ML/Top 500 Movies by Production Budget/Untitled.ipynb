# Top 500 movies by their budget 

Production budgets are not inflation adjusted.


Production budgets and box office numbers are hard to pin down specifically, as noted by The Numbers.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/top-500-movies.csv", parse_dates=["release_date"])
```

```python
plt.rcParams["figure.figsize"] = (14, 8)
sns.set_theme(style="whitegrid")
sns.set_palette(sns.color_palette("pastel"))
```

```python
data.head()
```

```python
plt.title("Opening weekend vs worldwide gross")
sns.scatterplot(data=data, x="opening_weekend", y="worldwide_gross", hue=data["release_date"].dt.year)
plt.xlabel("Opening Weekend")
plt.ylabel("Worldwide Gross");
```

```python
plt.title("Release date vs Production cost")

sns.barplot(data=data, x=data["release_date"].dt.year, y="production_cost")
plt.xticks(rotation=90);
```

```python
fig, (ax_1, ax_2) = plt.subplots(ncols=2, nrows=1)
sns.barplot(x=data["genre"], y=data["worldwide_gross"], ax=ax_1)
sns.barplot(x=data["mpaa"], y=data["worldwide_gross"], ax=ax_2)

ax_1.tick_params(axis="x", rotation=45)
plt.tight_layout()
```
