```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y);
```

```python
# First method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Creates axes
plt.show();
```

```python
# Second method
fig = plt.figure() # Create a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show();
```

```python
# Third method ( Recommended )
fig, ax = plt.subplots()
ax.plot(x, y); # Add some data
```

# Matplotlib example workflor

```python
# 0. Import matplotlib and get it ready for plotting in jupyter
import matplotlib.pyplot as plt
import os

# 1. Prepare the data
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]

# 2. Setup the plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot the data
ax.plot(x, y);

# 4. Customize plot
ax.set(title="Simple plot", 
      xlabel="x-axis",
      ylabel="y-label")

# 5. Save and show the figure
out = os.getcwd()
fig.savefig("{}/introduction-to-matplotlib".format(out))
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

colors = np.random.rand(1000)

x = np.random.rand(1, 1000)
y = np.random.rand(1, 1000)

ax.set(xlabel="Random x",
      ylabel="Randomt y",
      xlim=(0,0.5),
      ylim=(0,0.5))

ax.grid()
ax.scatter(x, y, c=colors, alpha=0.4)
```
