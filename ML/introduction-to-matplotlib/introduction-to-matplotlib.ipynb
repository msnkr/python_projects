# Matplotlib example workflor

```python
# 0. Import matplotlib and get it ready for plotting in jupyter
import matplotlib.pyplot as plt
import os

# 1. Prepare the data
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]

# 2. Setup the plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot the data
ax.plot(x, y);

# 4. Customize plot
ax.set(title="Simple plot", 
      xlabel="x-axis",
      ylabel="y-label")

# 5. Save and show the figure
out = os.getcwd()
fig.savefig("{}/introduction-to-matplotlib".format(out))
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

colors = np.random.rand(1000)

x = np.random.rand(1, 1000)
y = np.random.rand(1, 1000)

ax.set(xlabel="Random x",
      ylabel="Randomt y",
      xlim=(0,0.5),
      ylim=(0,0.5))

ax.grid()
ax.scatter(x, y, c=colors, alpha=0.4);
```

# Making figures with numpy arrays
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:5]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x ** 2);
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {
    "Almond butter": 10,
    "Peanut butter": 8,
    "Cashew butter": 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut butter store", xlabel="Price", ylabel="Product");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplot

```python
# Sublot option 1 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5));

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

N = np.random.randn(100)
# Plot to each different axis using list index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.randn(100), np.random.randn(100), c=N);
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));



# fig.savefig("subplots.png")
```
