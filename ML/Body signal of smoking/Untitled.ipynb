## About Dataset
### dataset
This dataset is a collection of basic health biological signal data.
The goal is to determine the presence or absence of smoking through bio-signals.

### The dataset is divided into type.

* ID : index
* gender
* age : 5-years gap
* height(cm)
* weight(kg)
* waist(cm) : Waist circumference length
* eyesight(left)
* eyesight(right)
* hearing(left)
* hearing(right)
* systolic : Blood pressure
* relaxation : Blood pressure
* fasting blood sugar
* Cholesterol : total
* triglyceride
* HDL : cholesterol type
* LDL : cholesterol type
* hemoglobin
* Urine protein
* serum creatinine
* AST : glutamic oxaloacetic transaminase type
* ALT : glutamic oxaloacetic transaminase type
* Gtp : γ-GTP
* oral : Oral Examination status
* dental caries
* tartar : tartar status
* smoking

### A higher-than-normal hemoglobin count means that your body is being deprived of oxygen

* Normal hemogoblin in males: 13.5 - 18g
* Normal hemogoblin in females: 12 – 15g

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv("../data/Body signal of smoking.csv")
df.head()
```

```python
df.gender.value_counts().plot(kind="bar", color=["lightblue", "salmon"])
plt.xticks(rotation=0);
# A lot more men in the dataframe
```

```python
pd.crosstab(df.age, df.gender).plot(kind="bar", color=["salmon", "lightblue"])
plt.xticks(rotation=45)
plt.legend(["Female", "Male"])
plt.xlabel("Age")
plt.ylabel("Amount")
plt.title("Age and Sex of DataFrame");
```

```python
pd.crosstab(df.age[df.smoking==1], df.gender[df.smoking==1]).plot(kind="bar", color=["salmon", "lightblue"])
plt.xticks(rotation=45)
plt.legend(["Female", "Male"])
plt.xlabel("Age")
plt.ylabel("Amount")
plt.title("Age and Sex of smokers in DataFrame");
# Men are defintely more likely to smoke. It slowly reduces over time or they dead. lol
```
