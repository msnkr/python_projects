

# Health Biological Signal Dataset

## Overview

The **Health Biological Signal Dataset** contains basic health-related biological signal measurements. The primary goal is to determine whether an individual is a smoker or a non-smoker based on these bio-signals.

### Dataset Details

- **Total Records**: 55,692
- **Number of Columns**: 27

## Column Descriptions

1. **ID**: Index representing each record.
2. **Gender**: Gender of the individual (e.g., male, female).
3. **Age**: Age group with a 5-year gap (e.g., 20-24, 25-29, etc.).
4. **Height (cm)**: Height in centimeters.
5. **Weight (kg)**: Weight in kilograms.
6. **Waist (cm)**: Waist circumference length.
7. **Eyesight (left)**: Eyesight measurement for the left eye.
8. **Eyesight (right)**: Eyesight measurement for the right eye.
9. **Hearing (left)**: Hearing assessment for the left ear.
10. **Hearing (right)**: Hearing assessment for the right ear.
11. **Systolic**: Blood pressure (systolic reading).
12. **Relaxation**: Blood pressure (diastolic reading).
13. **Fasting Blood Sugar**: Fasting blood sugar level.
14. **Cholesterol (total)**: Total cholesterol level.
15. **Triglyceride**: Triglyceride level.
16. **HDL (High-Density Lipoprotein)**: Cholesterol type (good cholesterol).
17. **LDL (Low-Density Lipoprotein)**: Cholesterol type (bad cholesterol).
18. **Hemoglobin**: Hemoglobin level.
19. **Urine Protein**: Presence of protein in urine.
20. **Serum Creatinine**: Serum creatinine level.
21. **AST (Glutamic Oxaloacetic Transaminase)**: Enzyme level in blood.
22. **ALT (Glutamic Pyruvic Transaminase)**: Enzyme level in blood.
23. **GTP (Î³-Glutamyltransferase)**: Enzyme level in blood.
24. **Oral Examination Status**: Status of oral health examination.
25. **Dental Caries**: Presence of dental caries (cavities).
26. **Tartar Status**: Presence of tartar on teeth.
27. **Smoking**: Indicates whether the individual is a smoker or non-smoker.

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_palette(sns.color_palette("pastel"))
plt.rcParams["figure.figsize"] = [14, 8]
```

```python
df = pd.read_csv("../data/Body signal of smoking.csv")
df.head()
```

```python
df.info()
```

```python
enc = LabelEncoder()

cat_features = ["gender", "oral", "tartar"]
for i in cat_features:
    df[i] = enc.fit_transform(df[i])
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(), linewidths=0.9);
```

```python
sns.histplot(data=df, x="age", kde=True, binwidth=5, hue="smoking");
```
