

# Health Biological Signal Dataset

## Overview

The **Health Biological Signal Dataset** contains basic health-related biological signal measurements. The primary goal is to determine whether an individual is a smoker or a non-smoker based on these bio-signals.

### Dataset Details

- **Total Records**: 55,692
- **Number of Columns**: 27

## Column Descriptions

1. **ID**: Index representing each record.
2. **Gender**: Gender of the individual (e.g., male, female).
3. **Age**: Age group with a 5-year gap (e.g., 20-24, 25-29, etc.).
4. **Height (cm)**: Height in centimeters.
5. **Weight (kg)**: Weight in kilograms.
6. **Waist (cm)**: Waist circumference length.
7. **Eyesight (left)**: Eyesight measurement for the left eye.
8. **Eyesight (right)**: Eyesight measurement for the right eye.
9. **Hearing (left)**: Hearing assessment for the left ear.
10. **Hearing (right)**: Hearing assessment for the right ear.
11. **Systolic**: Blood pressure (systolic reading).
12. **Relaxation**: Blood pressure (diastolic reading).
13. **Fasting Blood Sugar**: Fasting blood sugar level.
14. **Cholesterol (total)**: Total cholesterol level.
15. **Triglyceride**: Triglyceride level.
16. **HDL (High-Density Lipoprotein)**: Cholesterol type (good cholesterol).
17. **LDL (Low-Density Lipoprotein)**: Cholesterol type (bad cholesterol).
18. **Hemoglobin**: Hemoglobin level.
19. **Urine Protein**: Presence of protein in urine.
20. **Serum Creatinine**: Serum creatinine level.
21. **AST (Glutamic Oxaloacetic Transaminase)**: Enzyme level in blood.
22. **ALT (Glutamic Pyruvic Transaminase)**: Enzyme level in blood.
23. **GTP (Î³-Glutamyltransferase)**: Enzyme level in blood.
24. **Oral Examination Status**: Status of oral health examination.
25. **Dental Caries**: Presence of dental caries (cavities).
26. **Tartar Status**: Presence of tartar on teeth.
27. **Smoking**: Indicates whether the individual is a smoker or non-smoker.

This dataset provides valuable insights into the relationship between bio-signals and smoking behavior. Researchers and health professionals can use it to explore patterns and make informed decisions related to smoking cessation and overall health. ðŸŒ¿ðŸš­

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_palette(sns.color_palette("pastel"))
plt.rcParams["figure.figsize"] = [14, 8]
```

```python
df = pd.read_csv("../data/Body signal of smoking.csv")
df.head()
```

```python
df.T
```

```python
df.info()
```

```python
enc = LabelEncoder()

cat_features = ["gender", "oral", "tartar"]
for i in cat_features:
    df[i] = enc.fit_transform(df[i])
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr(), linewidths=0.9);
```

```python
plt.title("Age vs Smoking")
sns.histplot(data=df, x="age", kde=True, binwidth=5, hue="smoking");
```

```python
plt.title("Height and Weight vs Smokers")
sns.scatterplot(data=df, x="height(cm)", y="weight(kg)", hue="smoking");
```

```python
plt.title("Cholesterol vs Hemogoblin")
sns.scatterplot(data=df, x="Cholesterol", y="hemoglobin", hue="smoking");
```

Smokers seem to have a higher Hemogoblin

```python
plt.title("Urine protein in smokers vs non-smokers")
sns.barplot(data=df, x="age", y="Urine protein", hue="smoking");
```

Smokers have a higher urine protein than non-smokers. 

```python
plt.title("Dental caries in smokers vs non-smokers")
sns.barplot(data=df, x="age", y="dental caries", hue="smoking");
```

Again, smokers are more likely to have dental caries

```python
plt.title("Weight and waist size in smokers and non-smokers")
sns.scatterplot(data=df, x="weight(kg)", y="waist(cm)", hue="smoking");
```

```python
sns.scatterplot(data=df, x="age", y="relaxation", hue="smoking");
```

Based on this dataset, smoking does not really make someone relaxed. 

Smokers generally have a higher fasting blood sugar

```python
plt.title("Age and Fasting Blood Sugar vs Smokers and Non-Smokers,")
sns.scatterplot(data=df, x="age", y="fasting blood sugar", hue="smoking");
```

```python
_, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax_1 = sns.scatterplot(data=df, x="Cholesterol", y="LDL", hue="smoking", ax=ax1)
ax_2 = sns.scatterplot(data=df, x="Cholesterol", y="HDL", hue="smoking", ax=ax2)

ax_1.set_title("Cholesterol vs Bad Cholesterol in Smokers and Non-Smokers")
ax_2.set_title("Cholesterol vs Good Cholesterol in Smokers and Non-Smokers");
```

```python
sns.scatterplot(data=df, x="weight(kg)", y="Cholesterol", hue="smoking");
```

```python
X = df.drop("smoking", axis=1)
y = df["smoking"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
bst = XGBClassifier().fit(X_train, y_train)
```

```python
bst.score(X_test, y_test)
```
