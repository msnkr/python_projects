# Dataset Description: Crimes Against Women in India (2001-2021)

This dataset provides a comprehensive overview of various crimes against women across different states in India from 2001 to 2021. The data is organized into several key columns, each representing a specific type of crime or demographic information. Below is a brief description of each column:

- **State**: The name of the state or union territory in India.
- **Year**: The year in which the data was recorded.
- **Rape**: The number of reported rape cases.
- **K&A (Kidnap and Assault)**: The number of reported cases of kidnapping and assault.
- **DD (Dowry Deaths)**: The number of reported dowry-related deaths.
- **AoW (Assault against Women)**: The number of reported cases of general assault against women.
- **AoM (Assault against Modesty of Women)**: The number of reported cases of assault specifically targeting the modesty of women.
- **DV (Domestic Violence)**: The number of reported cases of domestic violence.
- **WT (Women Trafficking)**: The number of reported cases of women trafficking.

This dataset can be used to analyze trends, identify patterns, and develop insights into the prevalence and distribution of crimes against women in India over the two-decade period. It serves as a crucial resource for researchers, policymakers, and activists working towards the safety and empowerment of women.


```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"] = (14, 8)
sns.set_palette(sns.color_palette("pastel"))
sns.set_theme(style="darkgrid")
```

```python
d = pd.read_csv("../data/CrimesOnWomenData.csv", parse_dates=["Year"])
```

```python
d["Year"] = d["Year"].dt.year
```

```python
d.head()
```

```python
d["Total"] = d[["Rape", "K&A", "DD", "AoW", "AoM", "DV", "WT"]].sum(axis=1)
```

```python
d.head()
```

```python
plt.title("Year vs Total amount of cases")
sns.barplot(x=d["Year"], y=d["Total"], hue=d["Year"], legend=False)
plt.xlabel("Year")
plt.ylabel("Amount");
```

```python
total_d = d.sort_values(by="Total", ascending=False)[:20]
```

```python
sns.barplot(x=total_d["State"], y=total_d["Total"], hue=total_d["State"], legend=False);
```

```python
features = total_d[['Rape', 'K&A', 'DD', 'AoM', 'AoW', 'DV', 'WT', 'Total']]
```

```python
features.columns
```

```python
total_d["State"].unique()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(16, 20))

axes = axes.flatten()

for i, c in enumerate(features):
    sns.barplot(x=total_d["State"], y=features[c], ax=axes[i], hue=total_d["State"], legend=False)
    axes[i].set_title(f"State vs {c}")
    axes[i].set_xticklabels(total_d["State"], rotation=45)

plt.tight_layout()
```

### Overview of Selected Indian States

This section provides an overview of the characteristics of selected Indian states, focusing on their rural and economic status, as well as their significance as tourist destinations.

#### Uttar Pradesh
- **Rural and Poor**: Yes, many parts of Uttar Pradesh are rural and face economic challenges.
- **Touristy**: Yes, it has significant tourist attractions like the Taj Mahal in Agra and Varanasi, a major pilgrimage site.

#### Tripura
- **Rural and Poor**: Yes, Tripura is largely rural and one of the less economically developed states in India.
- **Touristy**: It has some tourist spots like Ujjayanta Palace and Neermahal, but it's not a major tourist destination.

#### Rajasthan
- **Rural and Poor**: Parts of Rajasthan are rural and face economic challenges, especially in the desert regions.
- **Touristy**: Yes, it's a major tourist destination known for its palaces, forts, and desert landscapes, including Jaipur, Udaipur, and Jaisalmer.

#### Punjab
- **Rural and Poor**: Punjab has rural areas, but it is relatively more prosperous due to its agricultural economy.
- **Touristy**: Yes, it has tourist attractions like the Golden Temple in Amritsar.

#### West Bengal
- **Rural and Poor**: Yes, there are rural and economically challenged areas, especially in the northern parts.
- **Touristy**: Yes, Kolkata is a major cultural hub, and Darjeeling is a popular hill station.

#### Maharashtra
- **Rural and Poor**: While Maharashtra has rural areas, it is one of the more economically developed states.
- **Touristy**: Yes, it has major tourist attractions like Mumbai, Ajanta and Ellora Caves, and hill stations like Lonavala.

#### Uttarakhand
- **Rural and Poor**: Yes, many parts are rural and economically less developed.
- **Touristy**: Yes, it's known for its natural beauty and pilgrimage sites like Rishikesh and Haridwar.

#### Madhya Pradesh
- **Rural and Poor**: Yes, it has significant rural areas and economic challenges.
- **Touristy**: Yes, it has tourist attractions like Khajuraho temples and national parks.

#### Andhra Pradesh
- **Rural and Poor**: Yes, there are rural areas and economic challenges, especially in the interior regions.
- **Touristy**: Yes, it has tourist spots like Tirupati temple and beaches in Visakhapatnam.
