```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/jobs_in_data.csv")
df.head()
```

```python
df.info()
```

```python
len(df)
```

```python
_, ax1 = plt.subplots(nrows=1, ncols=1)
ax1_bar = ax1.bar(df["work_year"], df["salary_in_usd"])
```

```python
_, ax1 = plt.subplots(nrows=1, ncols=1)
ax1_bar = ax1.bar(df["experience_level"], df["salary_in_usd"], color="red")
```

```python
plt.barh(df["job_category"], df["salary_in_usd"], color="orange");
```

```python
df.columns
```

```python
plt.figure(figsize=(12, 6))
job_title = df["job_category"].value_counts()
job_title.plot(kind="bar", color="purple");
```

```python
plt.figure(figsize=(14, 6))
df["employee_residence"].value_counts().plot(kind="bar");
```

```python
americans = df[df["employee_residence"] == "United States"]
```

```python
plt.figure(figsize=(12, 6))
americans["job_category"].value_counts().plot(kind="bar", color="green");
```

```python
everyone_else = df[df["employee_residence"] != "United States"]
```

```python
plt.figure(figsize=(14, 6))
everyone_else["employee_residence"].value_counts().plot(kind="bar", color="yellow");
```

```python
_, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(12, 6))
ax1_bar = ax1.bar(everyone_else["employee_residence"], everyone_else["salary_in_usd"])
ax1.legend(*ax1_bar.legend_elements());
```
