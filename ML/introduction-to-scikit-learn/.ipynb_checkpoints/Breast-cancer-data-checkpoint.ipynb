```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
# Get the data

breast = load_breast_cancer()
breast_cancer = pd.DataFrame(breast["data"], columns=breast["feature_names"])
breast_cancer["target"] = breast["target"]

breast_cancer.head()
```

```python
# Sort the data
X = breast_cancer.drop("target", axis=1)
y = breast_cancer["target"]
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
# Instantiate the model
clf = RandomForestClassifier().fit(X_train, y_train)
```

```python
# Score
# Default mean accuracy
clf.score(X_test, y_test)
```

```python
# Evaluate different metrics
y_preds = clf.predict(X_test)
y_preds[:10]
```

```python
print("Accuracy Score: {}".format(accuracy_score(y_test, y_preds)))
print("Precision Score: {}".format(precision_score(y_test, y_preds)))
print("Recall Score: {}".format(recall_score(y_test, y_preds)))
print("F1 Score: {}".format(f1_score(y_test, y_preds)))
```

```python
cross_val_score(clf, X, y)
```

```python
np.mean(cross_val_score(clf, X, y))
```
