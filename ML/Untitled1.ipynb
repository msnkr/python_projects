```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

I love the hex plots

```python
sns.jointplot(x=tips.tip, y=tips.total_bill, palette="pastel", kind="hex", color="Salmon");
```

```python
sns.jointplot(x=tips.tip, y=tips.total_bill, hue=tips.smoker, kind="kde", fill=True, palette="pastel");
```

```python
sns.scatterplot(x=tips.tip, y=tips.total_bill, hue=tips.day, palette="pastel");
```

```python
sns.histplot(x=tips.total_bill, hue=tips.sex, palette="pastel");
```
