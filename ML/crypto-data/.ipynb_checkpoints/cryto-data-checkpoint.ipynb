```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get data
data = pd.read_csv("BTC-USD.csv")
data.head()
```

```python
# Get the data over 2020
data_over_2020 = data[data['Date'].dt.year > 2020]
```

```python
plt.style.available
```

```python
# Plot the data for high and low
plt.style.use("seaborn-v0_8-whitegrid")
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1  ,figsize=(10, 6), sharex=True)

ax_1 = ax1.plot(data_over_2020["Date"], data_over_2020["High"], c="red")
ax_12 = ax1.plot(data_over_2020["Date"], data_over_2020["Low"], c="green");

ax2.plot(data_over_2020["Date"], data_over_2020["Open"], c="red")
ax2.plot(data_over_2020["Date"], data_over_2020["Close"], c="green")

ax1.set(title="High vs Low")
ax2.set(title="Open vs Close")

ax1.axhline(data_over_2020["High"].mean(), linestyle="--", c="red")
ax1.axhline(data_over_2020["Low"].mean(), linestyle="--", c="green")

ax2.axhline(data_over_2020["High"].mean(), linestyle="--", c="red")
ax2.axhline(data_over_2020["Low"].mean(), linestyle="--", c="green");
```
