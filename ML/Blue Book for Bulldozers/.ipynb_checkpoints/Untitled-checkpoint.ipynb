This is another workflow for blue book for bulldozers. I will be trying to use the KNNImputer

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_log_error

from catboost import CatBoostRegressor
```

```python
df = pd.read_csv("../data/Blue Book for Bulldozers/TrainAndValid.csv", low_memory=False)
```

```python
for label in df:
    if df[label].isna().sum() > 1:
        print(label, df[label].isna().mean())
```

There are a lot of missing values. KNNImputer might not work at all. But lets try. Why not?

```python
df.head()
```

```python
df.drop("SalesID", axis=1, inplace=True)
```

```python
cat_labels = [c for c in df if not pd.api.types.is_numeric_dtype(df[c])]
num_labels = [c for c in df if pd.api.types.is_numeric_dtype(df[c]) and c != "SalePrice"]
```

```python
cat_imputer = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),
    ("ohe", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
num_imputer = Pipeline([
    ("imputer", KNNImputer())
])
```

```python
preprocessor = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_labels),
    ("num_imputer", num_imputer, num_labels)
], remainder="passthrough")
```

```python
model = Pipeline([
    ("preprocessor", preprocessor),
    ("model", CatBoostRegressor())
])
```

```python
X, y = df.drop("SalePrice", axis=1), df["SalePrice"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
%%time 
model.fit(X_train, y_train)
```

This took way tooooo long. I need to test it out with a GPU

```python
%%time 
model.score(X_test, y_test)
```

Let me try SimpleImputer

```python
simple_num = Pipeline([
    ("imputer", SimpleImputer(strategy="mean"))
])

preprocessor = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_labels),
    ("simple_num", simple_num, num_labels)
], remainder="passthrough")
```

```python
model = Pipeline([
    ("preprocessor", preprocessor),
    ("model", CatBoostRegressor(verbose=False))
])
```

```python
%%time 

model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

The preprocessor with KNNImputer took 15 minutes and achieved a lower score than simple imputer with a mean strategy. If we don't care about real values then SimpleImputer would be a better fit. If a real value is recommended then we can stick to KNNImputer. KNNImputer also did better on datasets that had less missing values. So if the missing value is low, we should include KNNImputer.
